```
data = str(uploaded['data-03-diabetes.csv'])
data = data.replace("\\n", ',')
data = data.split(',')
data[0] = '-0.294118'
data.pop()
for e in data:
  e = float(e)
```

```
SIZE = int(6831/9)
```

```
x = np.empty( ( SIZE, 8) )
y = np.empty( ( SIZE, 1) )
w = np.ones((8))
```

```
for i in range(SIZE):
  x[i] = np.array(data[i*9:i*9+8])
  y[i] = data[i*9+8]
```

```
def H(i):
  return 1/(1+np.exp(-np.sum(x[i]*w)))
```

```
def Cost():
  ret = 0.0
  for i in range(SIZE):
    ret += y[i]*math.log(H(i)) + (1-y[i])*math.log(1-H(i))
    
  return -ret/SIZE
```

```
def GradiantH(i, j):
  return (x[i][j]*np.exp(-np.sum(x[i]*w)))/np.square(1+np.exp(-np.sum(x[i]*w)))
```

```
def GradiantCost(i):
  ret = 0.0
  for j in range(SIZE):
    ret += y[j]*GradiantH(j, i)/H(j) - (1-y[j])*GradiantH(j, i)/(1-H(j))
  return -ret/SIZE
    
```

```
LEARNING_RATE = 1e-2
for num_learn in range(10000):
  for i in range(8):
    w[i] -= GradiantCost(i)*LEARNING_RATE
    
  if num_learn % 50 == 0:
    print("cost : ",Cost())
    print("w : ", w)
    
```

```
cnt = 0
for i in range(SIZE):
  if(H(i) > 0.5):
    cnt += y[i]
  else:
    cnt += 1 - y[i]
    
print("정확도 : ", cnt/SIZE)
```
